#### requests 라이브러리 사용하기
- HTTP 요청시 사용
- urllib.request(파이썬 기본라이브러리) 보다 더 직관적이고 간단한 방법 제공
- 디코딩 처리도 적절히 해준다
- json 처리가 편함

```python
import requests
```

```python
reponse = requests.get("https://www.naver.com")

print(reponse.status_code)
print(reponse.encoding)

print(reponse.text) # 이렇게만 해주면 코드를 크롤링해서 가져올수있다, urllib 보다 간단해짐
```

```python
# 세션 활성화
s = requests.Session()

reponse = s.get("https://www.naver.com") # 세션 이용해서 요청

print(reponse.text)
s.close()
```

```python
# 세션 + with => close() 생략가능
with requests.Session() as s:
    reponse = s.get("https://www.naver.com")
    print(reponse.text)
```

#### 다양한 http 메소드 지원
- get()
- post()
- put()
- delete()
- https://httpbin.org/ (http 메소드를 연습 해볼수있는 사이트, 연습용 서버라서 응답이 잘 안될수있다)

```python
# get()
url = "https://httpbin.org/get"
with requests.Session() as s:
    reponse = s.get(url)
    print(reponse.text)
```

```python
# post()
url = "https://httpbin.org/post"

# 폼에 삽입되는 데이터는 dict 구조로 만들면된다
data = {"name":"hong"}

with requests.Session() as s:
    reponse = s.post(url,data=data)
    print(reponse.text)

#  "form": {
#     "name": "hong"
#   }, 
```

```python
# put()
url = "https://httpbin.org/put"

# 폼에 삽입되는 데이터는 dict 구조로 만들면된다
data = {"name":"Kim"}

with requests.Session() as s:
    reponse = s.put(url,data=data)
    print(reponse.text)
```

```python
# delete()
url = "https://httpbin.org/delete"

with requests.Session() as s:
    reponse = s.delete(url)
    print(reponse.text)
```

```python
# 파라메터를 get() 으로 보내기
# ex) https://httpbin.org/get?name=hong
url = "https://httpbin.org/get"

# 보낼 파라메터를 dict 구조로 만든다
# payload = {"name":"hong","age":25}
payload = {"name":"hong","age":25,"key1":["value1","value2"]}

with requests.Session() as s:
    reponse = s.get(url,params=payload)
    # print(reponse.text)
    print(reponse.url)
```

```python
# requests 라이브러리 json 처리
# jsonplaceholder 사이트 이용

url = "https://jsonplaceholder.typicode.com/todos/1"
with requests.Session() as s:
    reponse = s.get(url)
    print(reponse.json())
    print(reponse.text)
    print(reponse.content) # binary 형식으로 리턴해준다
```

```python
url = "https://jsonplaceholder.typicode.com/users"
with requests.Session() as s:
    reponse = s.get(url)
    
    for row in reponse.json():
        for k,v in row.items():
            print(f"{k} : {v}")
        print()
```

```python
# fake_useragent 도 사용가능
from fake_useragent import UserAgent

url = "https://httpbin.org/get"

userAgent = UserAgent()
hesders = {"user-agent":userAgent.chrome}

payload = {"name":"hong","age":25,"key1":["value1","value2"]}

with requests.Session() as s:
    reponse = s.get(url,params=payload,headers=hesders)
    # print(reponse.text)
    print(reponse.url)
```

```python
# 다음 쇼핑 => 베스트100 => f12 => Fetch/XHR => dispprodbest~ => 우클릭 Open in new tab => 나온창 주소복사
url = "https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=GMP&durationDays=30&count=100&_=1716515965528"

userAgent = UserAgent()
hesders = {"user-agent":userAgent.chrome}

# 1 product_name price_max 형태로 출력해보기
# SSL 에러 뜨면은 사이트 쪽에서 크롤링을 막은경우 좀 headers 를 바꿔서 요청하면된다, 이래도 안되면 좀 기다렷다가 시도해보고 그래도 안되면 지금은 어쩔수없다
with requests.Session() as s:
    reponse = s.get(url,headers=hesders)

    for idx,item in enumerate(reponse.json(),1):
        print(idx,item["product_name"],item["price_max"])
```
