```python
# get()
url = "https://httpbin.org/get"
with requests.Session() as s:
    reponse = s.get(url)
    print(reponse.text)
```

```python
# post()
url = "https://httpbin.org/post"

# í¼ì— ì‚½ì…ë˜ëŠ” ë°ì´í„°ëŠ” dict êµ¬ì¡°ë¡œ ë§Œë“¤ë©´ëœë‹¤
data = {"name":"hong"}

with requests.Session() as s:
    reponse = s.post(url,data=data)
    print(reponse.text)

#  "form": {
#     "name": "hong"
#   }, 
```

```python
# put()
url = "https://httpbin.org/put"

# í¼ì— ì‚½ì…ë˜ëŠ” ë°ì´í„°ëŠ” dict êµ¬ì¡°ë¡œ ë§Œë“¤ë©´ëœë‹¤
data = {"name":"Kim"}

with requests.Session() as s:
    reponse = s.put(url,data=data)
    print(reponse.text)
```

```python
# delete()
url = "https://httpbin.org/delete"

with requests.Session() as s:
    reponse = s.delete(url)
    print(reponse.text)
```

```python
# íŒŒë¼ë©”í„°ë¥¼ get() ìœ¼ë¡œ ë³´ë‚´ê¸°
# ex) https://httpbin.org/get?name=hong
url = "https://httpbin.org/get"

# ë³´ë‚¼ íŒŒë¼ë©”í„°ë¥¼ dict êµ¬ì¡°ë¡œ ë§Œë“ ë‹¤
# payload = {"name":"hong","age":25}
payload = {"name":"hong","age":25,"key1":["value1","value2"]}

with requests.Session() as s:
    reponse = s.get(url,params=payload)
    # print(reponse.text)
    print(reponse.url)
```

```python
# requests ë¼ì´ë¸ŒëŸ¬ë¦¬ json ì²˜ë¦¬
# jsonplaceholder ì‚¬ì´íŠ¸ ì´ìš©

url = "https://jsonplaceholder.typicode.com/todos/1"
with requests.Session() as s:
    reponse = s.get(url)
    print(reponse.json())
    print(reponse.text)
    print(reponse.content) # binary í˜•ì‹ìœ¼ë¡œ ë¦¬í„´í•´ì¤€ë‹¤
```

```python
url = "https://jsonplaceholder.typicode.com/users"
with requests.Session() as s:
    reponse = s.get(url)
    
    for row in reponse.json():
        for k,v in row.items():
            print(f"{k} : {v}")
        print()
```

```python
# fake_useragent ë„ ì‚¬ìš©ê°€ëŠ¥
from fake_useragent import UserAgent

url = "https://httpbin.org/get"

userAgent = UserAgent()
hesders = {"user-agent":userAgent.chrome}

payload = {"name":"hong","age":25,"key1":["value1","value2"]}

with requests.Session() as s:
    reponse = s.get(url,params=payload,headers=hesders)
    # print(reponse.text)
    print(reponse.url)
```

```python
# ë‹¤ìŒ ì‡¼í•‘ => ë² ìŠ¤íŠ¸100 => f12 => Fetch/XHR => dispprodbest~ => ìš°í´ë¦­ Open in new tab => ë‚˜ì˜¨ì°½ ì£¼ì†Œë³µì‚¬
url = "https://shoppinghow.kakao.com/siso/p/api/bestRank/dispprodbest?vCateId=GMP&durationDays=30&count=100&_=1716515965528"

userAgent = UserAgent()
hesders = {"user-agent":userAgent.chrome}

# 1 product_name price_max í˜•íƒœë¡œ ì¶œë ¥í•´ë³´ê¸°
# SSL ì—ëŸ¬ ëœ¨ë©´ì€ ì‚¬ì´íŠ¸ ìª½ì—ì„œ í¬ë¡¤ë§ì„ ë§‰ì€ê²½ìš° ì¢€ headers ë¥¼ ë°”ê¿”ì„œ ìš”ì²­í•˜ë©´ëœë‹¤, ì´ë˜ë„ ì•ˆë˜ë©´ ì¢€ ê¸°ë‹¤ë ·ë‹¤ê°€ ì‹œë„í•´ë³´ê³  ê·¸ë˜ë„ ì•ˆë˜ë©´ ì§€ê¸ˆì€ ì–´ì©”ìˆ˜ì—†ë‹¤
with requests.Session() as s:
    reponse = s.get(url,headers=hesders)

    for idx,item in enumerate(reponse.json(),1):
        print(idx,item["product_name"],item["price_max"])
```
