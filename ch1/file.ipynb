```python
# score.txt 읽어서 평균과 합 출력
with open("./resource/score.txt","r") as f:
    list1 = f.readlines()
    sum1 = 0
    for num in list1:
        print(num,type(num), end="")
        sum1 += int(num)
print("\n======================")
print("점수 총합 : ",sum1)
print("점수 평균 : ",sum1/len(list1))
```

```python
score = []
with open("./resource/score.txt","r") as f:
    for num in f:
        # print(num,end="")
        score.append(int(num))

print(sum(score))
print(sum(score)/len(score))
```

```python
# score.txt 파일을 읽어서 총합과 평균을 구한후 구한값을 result.txt 에 작성
# result.txt
# 총합 : 520
# 평균 : 86.67
score = []
with open("./resource/score.txt","r") as f:
    for num in f:
        # print(num,end="")
        score.append(int(num))

sum1 = sum(score)
avg = float(sum(score)/len(score))

with open("./resource/result.txt","w",encoding="utf-8") as f:
    f.writelines(f"총합 : {sum1}\n")
    f.writelines("평균 : %.2f\n" % avg)
```

```python
# info.txt
# 차다, 170, 48
import random
strings = list("가나다라마바사아자차카타파하")
with open("./resource/info.txt","w",encoding="utf-8") as f:
    for i in range(500):
        name = random.choice(strings)+random.choice(strings)
        height = random.randrange(150,200)
        weight = random.randrange(50,100)
        f.writelines(f"{name}, {height}, {weight}\n")
```

```python
# info.txt 파일을 읽어서 BMI 지수 계산
# bmi = weight / (height / 100 ** 2)
# bmi >= 25 과체중, bmi >= 18.5 정상체중, 나머지 저체중

# 바카, 197, 81
# 이름 : 바카
# 키 : 197
# 몸무게 : 81
# BMI :
# 결과 : 정상체중

with open("./resource/info.txt","r",encoding="utf-8") as f:
    for row in f:
        print(row,end="")
        name, height, weight = row.strip().split(", ") # 공백제거후 자르기

        bmi = int(weight) / ((int(height) / 100) ** 2)
        result = ""
        if bmi >= 25:
            result = "과체중"
        elif bmi >= 18.5:
            result = "정상체중"
        else:
            result = "저체중"
        # print(f"이름 : {name}\n")
        # print(f"키 : {height}\n")
        # print(f"몸무게 : {weight}\n")
        print("\n".join(["이름 : {}","키 : {}","몸무게 : {}","BMI : {}","결과 : {}"]).format(name,height,weight,bmi,result))
        print()
```

```python
content = None

while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 중 원하는 메뉴 번호 입력 ==>"))

    if no == 1:
        # origin.txt 파일을 읽은후 content 변수에 담기
        # content 변수에 담은 값을 chr(ord(가) + 100) 변경 후 파일에 작성 : encryption.txt
        with open("./resource/origin.txt","r",encoding="utf-8") as f:
            content = f.read()
        
        with open("./resource/encryption.txt","w",encoding="utf-8") as f:
            for c in content:
                f.write(chr(ord(c)+100))
    elif no == 2:
        # encryption.txt 파일을 읽어온 후 원래의 문자로 출력
        with open("./resource/encryption.txt","r",encoding="utf-8") as f:
            content = f.read()
            for i in range(0,len(content)):
                print(chr(ord(content[i])-100),end="")
    else:
        break
```
