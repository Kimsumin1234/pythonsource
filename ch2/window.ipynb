```python
# as p 별칭부여
import pyautogui as p
```

```python
# 마우스 좌표 가져오기

pos = p.position()
print(pos.x,pos.y)
```

```python
# 마우스 이동
# moveTo(x좌표,y좌표,시간)

p.moveTo(100,100,0.5)
p.moveTo(200,200,0.5)
p.moveTo(300,300,0.5)
```

```python
# moveTo() : 절대좌표
# move() : 상대좌표 (현재위치를 기준으로 움직인다)

p.moveTo(300,300,0.25)
p.move(100,100,0.25)
print(p.position())
p.move(100,100,0.25)
```

```python
# File 에 마우스를 가져다 대서 좌표알아내기

p.sleep(3)
print(p.position())
```

```python
# 마우스 클릭

p.click(58,16,1)
```

```python
# clicks : 마우스 클릭수

p.sleep(1)
p.click(clicks=5)
```

```python
# button="right" : 마우스 우클릭

p.click(clicks=2, interval=2, button="right")
```

```python
# 더블클릭

p.doubleClick(487,142)
```

```python
# 마우스 우클릭 (좌표지정 가능)

p.rightClick()
```

```python
# 마우스 위치 같은걸 알고싶을때 사용 (XY position)

p.mouseInfo()
```

```python
# 그림판에 자동으로 선 그리기 (좌표는 mouseInfo()로 알아냄)

p.sleep(3)
p.moveTo(200,370,0.5)
p.mouseDown()
p.moveTo(350,500,0.5)
p.mouseUp()
```

```python
# 메모장 드래그 이동

p.sleep(3)
p.moveTo(164,79,0.5)
p.drag(100,0,0.25)
```

```python
# vscode 스크롤 이동
# 양수: 위로, 음수: 아래로

p.sleep(2)
p.scroll(2000)
p.scroll(-2000)
```

```python
# 윈도우창에 현재 활성화 되있는걸 찾아옴

active = p.getActiveWindow()
print(active)
print(active.title)
print(active.size)
print(active.left,active.top,active.bottom,active.right)
```

```python
# 윈도우에 떠있는 모든창 가져오기

for w in p.getAllWindows():
    print(w)
```

```python
# 메모장만 가져오기

for w in p.getWindowsWithTitle("새 텍스트 문서"):
    print(w)
```

```python
# 창 활성화, 최대화, 최소화, 원래크기

w = p.getWindowsWithTitle("제목 없음")[0] # 제목이 같은게 여러개면 처음껄 선택
if not w.isActive:
    w.activate() # 창 활성화

if not w.isMaximized:
    w.maximize() # 창 최대화

p.sleep(1)
# if not w.isMinimized:
#     w.minimize() # 창 최소화

w.restore() # 원래 크기로
```

```python
# 자동으로 키보드 입력
import pyperclip

w = p.getWindowsWithTitle("제목 없음")[0]

if not w.isActive:
    w.activate()

p.write("write")
p.write("pyautogui",interval=0.25)
# p.write("안녕하세요") 이 상태로는 한글 작성 안됨

# 복사 붙여넣기로 한글작성 해야한다
pyperclip.copy("안녕하세요") # pyperclip(파이썬 클립보드) 에 복사(ctrl+c)
p.hotkey("ctrl","v")
```

```python
# 함수로 작성

def word_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
if not w.isActive:
    w.activate()

word_write("한글작성 ")
word_write("반갑습니다 ")
word_write("Hello !!! ")
```

```python
# hotkey() : 여러개의 키 조합 사용가능

p.hotkey("ctrl","shift","esc") # 작업관리자 창 띄우기
```
